```python
import subprocess
```

```python
proc = subprocess.run("git ls-files --exclude-standard",
                        shell=True,
                        stdout=subprocess.PIPE)
fileList = proc.stdout.decode().split('\n')
print(fileList)
```

```python
proc = subprocess.run("git branch --show-current",
                        shell=True,
                        stdout=subprocess.PIPE)
branchName = proc.stdout.decode()
print(branchName)
```
